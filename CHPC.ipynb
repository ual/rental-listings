```python
import pandas as pd
import numpy as np
```

```python
rents = pd.read_hdf('data/rents.h5', 'rents')
```

```python
rents.head()
```

```python
CA = rents[rents['state']=='CA']
```

```python
CA.head()
```

```python
CA['year'] = CA['date'].year
```

```python
CA['year'] = pd.DatetimeIndex(CA['date']).year
CA['month'] = pd.DatetimeIndex(CA['date']).month
```

```python
CA.head()
```

```python
#grouped = df.groupby('Team')
#print grouped['Points'].agg([np.sum, np.mean, np.std])
CA3 = CA[CA['bedrooms']<4]
grouped = CA3.groupby(['bedrooms','year','month'])
grouped['rent'].agg([np.median,np.size]).unstack(level=0).to_csv('data/CA.csv')
```

```python
grouped = CA3.groupby(['county','bedrooms','year','month'])
grouped['rent'].agg([np.median,np.size]).unstack(level=1).to_csv('data/Counties.csv')
```
